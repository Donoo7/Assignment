```python
#1. Create a Pandas Series from a list of integers and perform basic operations such as addition,
#subtraction, multiplication, and division on the series.
import numpy as np
import pandas as pd
data = [-1,1,11,2,-3,6]
num = pd.Series(data)
add = num + 2
sub = num - 2
multi = num * 2
div = num / 2
print("addition = ",add,"\n",
     "substraction =",sub,"\n",
     "multiplication =",multi,"\n",
     "division =",div,"\n",)
```

```python
#3. Sort a Series in ascending and descending order based on values.
num_ascend = sorted(data1)
num_descend = sorted(data1, reverse=True)
print("ascending = ",num_ascend,"\n"
     "descending = ",num_descend,"\n")
```
